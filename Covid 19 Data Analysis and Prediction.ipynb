# Converting Date and time to ordinal

```python
# Importing datetime and converting date column to ordinal form
import datetime as dt
df.loc[:,'date'] = pd.to_datetime(df['date'])
df.loc[:,'date'] = df['date'].map(dt.datetime.toordinal)
```

# Data Filtering

```python
use_df=df.drop(['iso_code','continent','location','tests_units','new_deaths','total_deaths','total_deaths_per_million','new_deaths_per_million'],axis=1)
#after dropping we can see the basic details of the useful data
use_df.info()
```

```python
use_df.drop('total_cases_per_million',axis=1)#we are dropping the total cases per million so that we can 
#only predict the total cases and let not the the total cases per million affect our prediction model
```

```python
use_df.describe().transpose()
```

```python
use_df.shape
```

# Splitting the data into features and target data

```python
y=use_df['total_cases']
```

```python
X=use_df.drop(['total_cases','total_cases_per_million'],axis=1)#we are dropping 
```

# Performing train-test split 

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
print("testing data of features columns")
X_test.head(5)
```

```python
print("training data of target column")
y_train.head(5)
```

```python
print("testing data of target column")
y_test.head(5)
```

# Data Modelling

# Implementing Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
mod=LinearRegression()
mod.fit(X_train,y_train)
```

# Printing Accuracy of Linear Regression Model

```python
accuracy =mod.score(X_test,y_test)*100
print("the accuracy of model is: ",accuracy,"%")
```

# Implementing Linear Regression

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rfg=RandomForestRegressor()
rfg.fit(X_train,y_train)
```

# Printing Accuracy of Random Forest Regressor

```python
accu=rfg.score(X_test,y_test)*100
print("the accuracy of random forest generator: ",accu,"%")
```
